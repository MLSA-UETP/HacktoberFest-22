## b) 4 V's of Big Data
##### These Vs stand for the four dimensions of Big Data: Volume, Velocity, Variety and Veracity.

### VOLUME:
We produce the same quantity of data today, every minute, as we did from the beginning of time until the year 2000. The size of the data that needs to be processed is now described in terms of terabytes and petabytes. For data to qualify as big data, its size is unquestionably a crucial factor. Due to the volume of data we handle every day, new technologies and methods, such multitiered storage media, have been developed to safely gather, process, and store it.The sheer volume of the data requires distinct and different processing technologies than traditional storage and processing capabilities. In other words, this means that the data sets in Big Data are too large to process with a regular laptop or desktop processor. An example of a high-volume data set would be all credit card transactions on a day within Europe
### VELOCITY:
The pace at which fresh data is created and disseminated is what is meant by the term "velocity." Sending an SMS, looking at your social media feed, responding to posts on Facebook, Instagram, or Twitter, or using your credit card to make a purchase all produce data that needs to be processed immediately. When you multiply these actions by the number of people doing the same and more throughout the world, you can begin to understand why velocity is an important characteristic of big data.
### VARIETY:
Unstructured, semi-structured, and structured data are the three main categories of data available today. The algorithms needed to process the different types of produced data vary depending on the data that has to be processed. In the past, data was well-structured; consider relational databases or Excel spreadsheets. Big data's ability to contain unstructured data that is difficult to organise into a spreadsheet, as well as structured data like text, photos, videos, and voice recordings, is one of its important characteristics. The rules that apply to structured data do not apply to unstructured data. Today, we can employ technology to make sense of unstructured data in a manner that wasn't previously feasible.
### VERACITY:
Big data's veracity indicates how reliable the data is. Is the data precise and of a high calibre? To be able to draw appropriate conclusions from large data that originates from a number of sources, it's critical to comprehend the chain of custody, the metadata, and the context at the time the data was acquired. The more reliable the data, the more significant it is to assess and contribute to meaningful outcomes for a company.

## c) Structure of Data
##### Big data structures can be divided into three categories â€“ structured, unstructured, and semi-structured. 

### Structured:
It is the data that adheres to a predetermined structure and is easy to evaluate. It follows a tabular pattern with connections between distinct rows and columns. Consider SQL databases as a popular illustration. Structured data depends on how data may be handled, retrieved, and stored. It is regarded as the most "conventional" kind of data storage.
### UnStructured:
This kind of huge data has an ambiguous shape, making it impossible to store it in a conventional manner and evaluate it without first converting it into a structured format. Multimedia resources such as audio, video, and photographs can be used as instances of unstructured data. It's critical to realise that unstructured data is now expanding at a quicker rate than other big data kinds.
### Semi-Structured:
It is a kind of large data that doesn't follow a prescribed framework for data models. However, some of the organisational tags or other markers that are included with the data serve to enforce field and record hierarchies and to distinguish semantic aspects. This kind of massive data can be compared to XML or JSON files. This category was created because semi-structured data is much simpler to evaluate than unstructured data. A large variety of big data tools and solutions can read and interpret XML files or JSON documents, which lessens the complexity of the analysis process.

#### Dealing with missing values

Firstly, we obtain the number of missing data per column, then we found total missing valus and from that we obtain the percentage of missing data.

```python
# get the number of missing data points per column
missing_values_count=crime_data.isnull().sum()
missing_values_count
```

```python
# how many total missing values do we have?
total_cells = np.product(crime_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

#### Dealing with missing data

We replaced all the missing data with the value that comes directly after it in same column and replaced rest of our data with with 0's

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
crime_data = crime_data.fillna(method='bfill', axis=0).fillna(0)
```

#### RELATIONSHIP OF VARIABLES

Let us now examine the relationships of variables with eachother

We made a heatmap to observe the coorelation of each attribute.

Each square shows the correlation between the variables on each axis. Correlation ranges from -1 to +1. Values closer to zero means there is no linear trend between the two variables. The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increases. 



```python
##heatmap of correlation of the data
sns.heatmap(crime_data.corr(),linewidths=0.1,vmax=1.0, square=True, linecolor='white', annot=True)
```
